```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('base_final.csv')
```

```python
year = pd.Categorical(df["ao"])
```

```python
data = df.set_index(["edo","ao"])
```

```python
data['ao'] = year
```

```python
print(data.head())
```

```python
data["logpibe"] = np.log(data["pibe"])
data["prop25a34"] = data["edad25a34"]*100/data["pob_tot"]
```

```python
import linearmodels as linearmodels
from linearmodels.panel import PooledOLS
import statsmodels.api as sm
```

```python
exog_vars = ['prop25a34','desocupa','logpibe','pobreza']
exog = sm.add_constant(data[exog_vars])
mod = PooledOLS(data.tasa_rcasa, exog)
pooled_res = mod.fit()
print(pooled_res)
```

```python
from linearmodels.panel import RandomEffects
mod = RandomEffects(data.tasa_rcasa, exog)
re_res = mod.fit()
print(re_res)
```
